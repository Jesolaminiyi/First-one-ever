```python
# Checking the data type to see if its in the right format for analysis

car_dataset.dtypes
```

```python
# Cleaning and Manipulating the data

car_dataset["body"] = car_dataset["body"].astype(str).apply(change_now)

car_dataset["color"] = car_dataset["color"].astype(str).apply(change)

car_dataset["make"] = car_dataset["make"].astype(str).apply(change_now)

car_dataset["Month"] = car_dataset["saledate"].astype(str).apply(month).astype(str)

car_dataset["Month"] = car_dataset["Month"].astype(str).apply(func)

car_dataset["week_day"] = car_dataset["saledate"].astype(str).apply(week)

car_dataset["saledate"] = car_dataset["saledate"].astype(str)

car_dataset["week_day"] = car_dataset["week_day"].apply(change)

car_dataset["year"] = car_dataset["year"].astype(str)
```

```python
# Checking the data type to see if its in the right format for analysis

car_dataset.dtypes
```

```python
# Plotting my First Chart to see What color of cars is mostly purchased 

car_dataset["color"].value_counts().plot(kind="bar", figsize = (20,5), title = "TOTAL CAR PUCHASED BY COLOR", ylabel="Count Of Purchase",xlabel = "Car Color", rot = "horizontal")
```

```python
# The second chart was to see how much cars is being purchased per Car Make 

car_dataset["make"].value_counts().head(20).plot(kind = "barh", figsize= (10,20), ylabel = "Car Make", title = "COUNT OF CAR PURCHASE PER CAR_MAKE")
plt.xlabel("Count Of Car Purchase", size = 12)
plt.ylabel("Car Make",size = 12)
plt.title("COUNT OF CAR PURCHASE PER CAR_MAKE", size = 14)
plt.show()
```

```python
# Checking to see What size is mostly purchased

car_dataset["body"].value_counts().head(20).plot(kind = "barh",figsize = (10,20))
plt.title("COUNT OF CAR PURCHASE PER TOP 20 BODY TYPE")
plt.ylabel("Car Body Type")
plt.xlabel("Count_Of_Purchase")
plt.show()
```

```python
# Checking to see the years with the most purchase


car_dataset["year"].value_counts(ascending= True).plot(kind = "bar", figsize = (15,5), xlabel = "Year Of Purchase", ylabel = "Count Of Purchase", title = "Total Purchase by year")
```

```python
# The weekdays with the most purchases

car_dataset["week_day"].value_counts().plot(kind = "bar", title = "Total Purchase by Week_days", ylabel= "Count Of Purchase")
```

```python
# Checking to See the Month with The most Purchase over the years

pivot2 = car_dataset.pivot_table(index = "Month", values="make", aggfunc= "count")

pivot2.plot(kind = "bar", title ="TOTAL PURCHASE BY MONTH", ylabel= "Count Of Purchase")
```

```python
%matplotlib inline

# Looking into if the price affects the count of Purchase or Total purchase of different car(body size)


one = car_dataset.groupby("body")["sellingprice"].sum()
two = car_dataset.groupby("body")["sellingprice"].median()


fig, ax1 = plt.subplots(figsize = (20,10))
ax2 = ax1.twinx()
ax1.bar(one.index,one, color = "g")
ax2.plot(one.index,two, "-o")
ax1.set_xticklabels(one.index, rotation = "vertical", size = 8)
ax1.set_title("Count_Of_Purchase_By_Price")
plt.show()
```
